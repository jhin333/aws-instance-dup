```python
# Must be included at the beginning of each new notebook. Remember to change the app name.
import findspark
findspark.init('/home/ubuntu/spark-3.2.1-bin-hadoop2.7')
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('a3').getOrCreate()
# Schemas can only be inferred for CSV files. 
df = spark.read.csv('COVID DEATHS.csv', inferSchema=True, header=True)
df.show()
```

```python
# 转换为 Pandas DataFrame
pandas_df = df.toPandas()

# 导入绘图库
import matplotlib.pyplot as plt

# 选择你要绘制的列，此处假设列名为 'columnName'
column_name = 'total_deaths'

# 绘制直方图
pandas_df[column_name].plot(kind='hist', rwidth=0.8)
plt.xlabel(column_name)
plt.ylabel('Frequency')
plt.title(f'Histogram of {column_name}')

# 显示图形
plt.show()
```

```python
import seaborn as sns
# 使用 seaborn 和 matplotlib 绘制直方图
plt.figure(figsize=(10,6))
sns.histplot(data=pandas_df, x='new_deaths', bins=30, kde=True)
plt.xlabel('New Deaths')
plt.ylabel('Frequency')
plt.title('Histogram of New Deaths')
plt.show()
```

```python
# 转换为 Pandas DataFrame
pandas_df = df.toPandas()

# 导入绘图库
import matplotlib.pyplot as plt

# 选择你要绘制的列，此处假设列名为 'xColumn' 和 'yColumn'
x_column = 'total_deaths'
y_column = 'total_vaccinations'

# 使用 Matplotlib 绘制散点图
plt.figure(figsize=(10, 6))
plt.scatter(pandas_df[x_column], pandas_df[y_column], alpha=0.5)
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.title(f'Scatter Plot of {y_column} vs {x_column}')

# 显示图形
plt.show()
```

```python
import findspark
findspark.init('/home/ubuntu/spark-3.2.1-bin-hadoop2.7')

import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, when
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# 使用 PySpark 创建一个新列 'total_cases_bins' 来对 'total_cases' 进行分箱
bins = [0, 1000, 10000, 100000, 1000000, 10000000]
bin_labels = ['0-1k', '1k-10k', '10k-100k', '100k-1M', '1M-10M']

df = df.withColumn(
    'total_cases_bins',
    when(col('total_cases') < 1000, bin_labels[0]).
    when(col('total_cases') < 10000, bin_labels[1]).
    when(col('total_cases') < 100000, bin_labels[2]).
    when(col('total_cases') < 1000000, bin_labels[3]).
    otherwise(bin_labels[4])
)

# 转换为 Pandas DataFrame
pandas_df = df.toPandas()

# 使用 seaborn 和 matplotlib 绘制箱线图
plt.figure(figsize=(10,6))
sns.boxplot(x='total_cases_bins', y='total_deaths', data=pandas_df, order=bin_labels)
plt.xlabel('Total Cases')
plt.ylabel('Total Deaths')
plt.title('Boxplot of Total Deaths vs Total Cases')
plt.show()

```
