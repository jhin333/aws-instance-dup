```python
import findspark
findspark.init('/home/ubuntu/spark-3.2.1-bin-hadoop2.7')

import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, when
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Use PySpark to create a new column 'total_cases_bins' to bin 'total_cases'
bins = [0, 1000, 10000, 100000, 1000000, 10000000]
bin_labels = ['0-1k', '1k-10k', '10k-100k', '100k-1M', '1M-10M']

df = df.withColumn(
    'total_cases_bins',
    when(col('total_cases') < 1000, bin_labels[0]).
    when(col('total_cases') < 10000, bin_labels[1]).
    when(col('total_cases') < 100000, bin_labels[2]).
    when(col('total_cases') < 1000000, bin_labels[3]).
    otherwise(bin_labels[4])
)

pandas_df = df.toPandas()

plt.figure(figsize=(10,6))
sns.boxplot(x='total_cases_bins', y='total_deaths', data=pandas_df, order=bin_labels)
plt.xlabel('Total Cases')
plt.ylabel('Total Deaths')
plt.title('Boxplot of Total Deaths vs Total Cases')
plt.show()
```

```python
# Calculate the median of each column
columns_to_fill = ['total_cases', 'new_cases', 'total_deaths', 'total_vaccinations']
medians = {column: df.approxQuantile(column, [0.5], 0)[0] for column in columns_to_fill}

# Fill missing values with median
df= df.na.fill(medians)

mean_value = df.select(mean(col("new_deaths"))).collect()[0][0]

df = df.withColumn("new_deaths", when(col("new_deaths").isNull(), mean_value)\
                                        .otherwise(col("new_deaths")))

columns_to_check = ['total_cases', 'new_cases', 'total_deaths', 'new_deaths','total_vaccinations']


missing_values = df.select([sum(col(c).isNull().cast('int')).alias(c) 
                                   for c in columns_to_check])

missing_values.show()

selected_columns = df.select('total_cases', 'new_cases', 'total_deaths', 'new_deaths','total_vaccinations')

selected_columns.show()
df.select('new_deaths').describe().show()
```

```python
from pyspark.sql.functions import col, when

columns_to_check = ['total_cases', 'new_cases', 'total_deaths', 'total_vaccinations']

# Handle outliers in each column
for column in columns_to_check:
    # Calculate Q1 and Q3
    Q1, Q3 = df.approxQuantile(column, [0.25, 0.75], 0)
    IQR = Q3 - Q1
    
    # Define the boundaries of outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Handle outliers - choose to replace with boundary values or optionally remove outliers
    df = df.withColumn(column,
                       when(col(column) < lower_bound, lower_bound)
                       .when(col(column) > upper_bound, upper_bound)
                       .otherwise(col(column)))
    
    outliers = df.filter((col(column) < lower_bound) | (col(column) > upper_bound)).select(column)
  
    if outliers.count() > 0:
        print(f"Outliers for {column}:")
        outliers.show()
    else:
        print(f"No outliers detected for {column}.")
```

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.stat import Correlation
from pyspark.sql.functions import col
from pyspark.sql import SparkSession
from pyspark.ml.classification import LogisticRegression

# DataFrame
numerical_cols = ["total_deaths", "population", "total_cases", "new_cases", "new_deaths", "total_vaccinations"]

# Use VectorAssembler to combine numeric features into a vector
assembler = VectorAssembler(inputCols=numerical_cols, outputCol='features')
numerical_merged_df = assembler.transform(df_cleaned)

# Calculate correlation coefficient matrix
r1 = Correlation.corr(numerical_merged_df, 'features').head()
corr_matrix = r1[0].toArray()

# Print the correlation coefficient of 'total_deaths' with other features
for i, col in enumerate(numerical_cols):
    print(f"Correlation between total_deaths and {col}: {corr_matrix[i][numerical_cols.index('total_deaths')]}")
    

# DataFrame
target_col = 'total_deaths'
feature_cols = [col for col in df_cleaned.columns if col != target_col]

# Functions for feature selection using RFE
def rfe(df_cleaned, feature_cols, target_col, num_features_to_select):
    selected_features = feature_cols.copy()
    
    while len(selected_features) > num_features_to_select:
        # Combine features into a vector
        assembler = VectorAssembler(inputCols=selected_features, outputCol='features')
        df_cleaned = assembler.transform(df_cleaned)

        # Train a logistic regression model
        lr = LogisticRegression(featuresCol='features', labelCol=target_col)
        model = lr.fit(df_cleaned)
        
        # Get the coefficients of the feature
        coefficients = model.coefficients.toArray()

        # Find the feature with the smallest absolute value of the coefficient
        min_coeff_idx = min(range(len(coefficients)), key=lambda i: abs(coefficients[i]))

        # Remove this feature
        feature_to_remove = selected_features[min_coeff_idx]
        selected_features.remove(feature_to_remove)
        
        print(f'Removed feature: {feature_to_remove}')

    return selected_features

# Call the RFE function for feature selection
selected_features = rfe(df_cleaned, feature_cols, target_col, 5)

print('Selected features:', selected_features)
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import log, col
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression
from pyspark.ml.evaluation import RegressionEvaluator

selected_columns = ["total_deaths", "population", "total_cases", "new_cases", "new_deaths", "total_vaccinations"]
df_cleaned = df_cleaned.select(selected_columns)


for col_name in ["total_cases", "new_cases", "new_deaths", "total_vaccinations", "population"]:
    df_cleaned = df_cleaned.withColumn(f"log_{col_name}", log(col(col_name) + 1))  # 


feature_columns = ["log_population", "log_total_cases", "log_new_cases", "log_new_deaths", "log_total_vaccinations"]
assembler = VectorAssembler(inputCols=feature_columns, outputCol="features")
df_features = assembler.transform(df_cleaned)


train_data, test_data = df_features.randomSplit([0.8, 0.2], seed=42)


lr = LinearRegression(featuresCol="features", labelCol="total_deaths")


lr_model = lr.fit(train_data)


predictions = lr_model.transform(test_data)

# Evaluate model performance
evaluator = RegressionEvaluator(labelCol="total_deaths", predictionCol="prediction", metricName="rmse")
rmse = evaluator.evaluate(predictions)
print(f"Root Mean Squared Error (RMSE): {rmse}")

# Output model coefficients
coefficients = lr_model.coefficients
intercept = lr_model.intercept
print(f"Coefficients: {coefficients}")
print(f"Intercept: {intercept}")
```

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression
import pandas as pd


selected_columns = ["total_deaths", "population", "total_cases", "new_cases", "new_deaths", "total_vaccinations"]
df_cleaned = df_cleaned.select(selected_columns)


feature_columns = ["population", "total_cases", "new_cases", "new_deaths", "total_vaccinations"]
vector_assembler = VectorAssembler(inputCols=feature_columns, outputCol="features")
df_assembled = vector_assembler.transform(df_cleaned)


df_final = df_assembled.select("total_deaths", "features")


train_data, test_data = df_final.randomSplit([0.8, 0.2], seed=123)


lr = LinearRegression(featuresCol="features", labelCol="total_deaths")


lr_model = lr.fit(train_data)


coefficients = lr_model.coefficients
intercept = lr_model.intercept


print("Coefficients: " + str(coefficients))
print("Intercept: " + str(intercept))


predictions = lr_model.transform(test_data)


predictions.select("total_deaths", "prediction").show()
```

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import DecisionTreeClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from sklearn.metrics import accuracy_score

#Select selected columns
selected_columns = ["total_deaths", "population", "total_cases", "new_cases", "new_deaths", "total_vaccinations"]
df_cleaned = df_cleaned.select(selected_columns)

#Combine feature columns using VectorAssembler
feature_columns = ["population", "total_cases", "new_cases", "new_deaths", "total_vaccinations"]
vector_assembler = VectorAssembler(inputCols=feature_columns, outputCol="features")
output = vector_assembler.transform(df_cleaned)

#Select data columns for training and testing
output = output.withColumn("label", (output["total_deaths"] > 1000).cast("double"))

final_data = output.select("features", "label")

#Split training and test data sets
train_data, test_data = final_data.randomSplit([0.85, 0.15], seed=123)

#Create and train a decision tree classifier
dt_classifier = DecisionTreeClassifier(labelCol="label", featuresCol="features")

dtModel_classifier = dt_classifier.fit(train_data)

#Use the model to make predictions
predictions = dtModel_classifier.transform(test_data)

#Calculation accuracy
evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="accuracy")
accuracy = evaluator.evaluate(predictions)
print("Accuracy: {}".format(accuracy))

#Print feature importance
feature_importances = dtModel_classifier.featureImportances
for feature, importance in zip(feature_columns, feature_importances):
    print("{} Importance: {}".format(feature, importance))
    


```

```python
feature_importances = dtModel_classifier.featureImportances

feature_columns = ["population", "total_cases", "new_cases", "new_deaths", "total_vaccinations"]

# create an empty list to store the importance value of each predictor variable
importances = []

# Add the importance value of each predictor variable to the importances list
for feature, importance in zip(feature_columns, feature_importances.toArray()):  
    print("{} Importance: {}".format(feature, importance))  
    importances.append(importance)


plt.bar(feature_columns, importances, color='blue')  
plt.xlabel('Features') 
plt.ylabel('Importance')  
plt.title('Feature Importances') 
plt.xticks(rotation=45)  
plt.show()  
```

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import DecisionTreeClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.mllib.evaluation import MulticlassMetrics
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 
predictions_and_labels = predictions.select("prediction", "label").rdd.map(tuple)

#  PySpark  MulticlassMetrics 
metrics = MulticlassMetrics(predictions_and_labels)
confusion_matrix = metrics.confusionMatrix().toArray()

print("Confusion Matrix:")
print(confusion_matrix)

#  seaborn  heatmap
plt.figure(figsize=(10, 7))
sns.set(font_scale=1.4)  # 
sns.heatmap(confusion_matrix, annot=True, fmt="g", cmap="Blues",
            xticklabels=["Predicted 0", "Predicted 1"], 
            yticklabels=["Actual 0", "Actual 1"])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
# Calculate F1 score
f1_evaluator = MulticlassClassificationEvaluator(metricName="f1")
f1 = f1_evaluator.evaluate(predictions)
print("F1 Score: {}".format(f1))

# Calculation accuracy
precision_evaluator = MulticlassClassificationEvaluator(metricName="weightedPrecision")
precision = precision_evaluator.evaluate(predictions)
print("Precision: {}".format(precision))

# Calculate recall
recall_evaluator = MulticlassClassificationEvaluator(metricName="weightedRecall")
recall = recall_evaluator.evaluate(predictions)
print("Recall: {}".format(recall))
```
